```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

df = pd.read_excel('https://www.eia.gov/electricity/data/state/annual_generation_state.xls')
df.columns = ["Year","State", "Type of Producer", "Energy Source", "Total Energy Generated(MWh)"]
df = df[df["Type of Producer"].str.contains("Total Electric Power Industry")]
df = df[df["Energy Source"].str.contains("Total|Hydroelectric Conventional|Wind|Solar Thermal and Photovoltaic")]
years = [2011,2012,2013,2014,2015,2016,2017,2018]
df = df[df["Year"].isin(years)]
df_total_energy_generated = df[df["Energy Source"].str.contains("Total")]
df_rewable_energy_generated = df[df["Energy Source"].str.contains("Hydroelectric Conventional|Wind|Solar Thermal and Photovoltaic|Pumped Storage")]
df_new = df_rewable_energy_generated.groupby(['Year','State']).sum()[["Total Energy Generated(MWh)"]]
df_new.reset_index()
df = df.set_index("Year")
df_total_energy_generated = df_total_energy_generated.set_index("Year")
df_new.reset_index(inplace=True)
df_new = df_new.set_index("Year")
df = df[df["State"].str.contains("CA|CO|MT|MN|NY|VT|VA|WA")]
df_new = df_new[df_new["State"].str.contains("CA|CO|MT|MN|NY|VT|VA|WA")]
df_total_energy_generated = df_total_energy_generated[df_total_energy_generated["State"].str.contains("CA|CO|MT|MN|NY|VT|VA|WA")]
df_new.reset_index(inplace = True)
df.reset_index(inplace = True)
df_total_energy_generated.reset_index(inplace = True)
df = df[df["Energy Source"].str.contains("Total")]
df = df.drop(df.columns[[2, 3]], axis=1)
df.reset_index(inplace = True)
df["Renewable Energy Generated(MWh)"] = df_new["Total Energy Generated(MWh)"]
df["Proportion"] = df_new["Total Energy Generated(MWh)"] / df_total_energy_generated["Total Energy Generated(MWh)"]

df.head(8)
```
