## Original Dataset:
https://www.eia.gov/environment/emissions/state/excel/electricity.xlsx

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("annual_electricity_related_CO2_emissions.csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[1]
```

```python
df.head()
```

```python
states = ["California", "Colorado", "Connecticut", "Florida", "Montana", "Michigan", "Minnesota", "New Jersey", "New York", "Oregon", "Tennessee", "Texas", "Vermont", "Virginia", "Washington", "Wisconsin"]
```

```python
new_df = df.loc[(df["State"].isin(states))]
```

```python
new_df
```

```python
remyears = []
for i in range(1980,2011):
    remyears.append(i)
```

```python
rem = []
for year in remyears:
    rem.append(str(year))
rem
```

```python
new_df.drop(columns=rem, inplace=True)
```

```python
new_df.drop(columns=["Percent", "Absolute"], inplace=True)
```

```python
new_df
```

```python
new_df.reset_index(drop=True, inplace=True)
```
